```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_excel("IncomeandExpensesdataset.xlsx")
dataset
```

```python
dataset.info()
```

```python
dataset.plot(kind='scatter', x='PENDAPATAN', y='PENGELUARAN')
plt.show()
```

```python
model = LinearRegression()
model.fit(dataset['PENDAPATAN'].values.reshape(-1, 1), dataset['PENGELUARAN'].values)
```

```python
garislinear = model.predict(dataset['PENDAPATAN'].values.reshape(-1, 1))
garislinear
```

```python
plt.plot(dataset['PENDAPATAN'], garislinear, color='r', label='Garis Linear')
plt.scatter(dataset['PENDAPATAN'], dataset['PENGELUARAN'], color='b', label='Data')
plt.xlabel('PENDAPATAN')
plt.ylabel('PENGELUARAN')
plt.legend()
plt.show()
```
